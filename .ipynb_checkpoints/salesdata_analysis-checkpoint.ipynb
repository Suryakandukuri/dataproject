```python
# importing required libraries
# 
import pandas as pd
import numpy as np
import glob
from pathlib import Path
import altair as alt
import datetime
```

```python
# combining data for all months
data_folder = Path("./Sales_data/")

filepath_list = Path(data_folder).glob("*.csv")

# dataframe_list
df_list = []

# iterating over all the files
for filepath in filepath_list:
    df = pd.read_csv(filepath, parse_dates=True, date_parser = True)
    df_list.append(df)

sales_data = pd.concat(df_list)

sales_data.head()
```

```python
# creating separate columns for address, city, state and pin

addr_df = sales_data["Purchase Address"].str.split(', ', expand=True)

addr_df = pd.concat([addr_df, addr_df[2].str.split(' ', expand=True)], axis=1)

addr_df.columns = ["address", "city", "statezip","state", "zip"]

addr_df = addr_df.drop(columns ="statezip")

addr_df
```

```python
# adding the address data to the original sales_data

sales_data = pd.concat([sales_data, addr_df], axis=1)
```

```python
# remove rows which have all null values

sales_data = sales_data.dropna(how = 'all')

# remove rows which have column names as rows
sales_data = sales_data[sales_data["Order ID"] != "Order ID"]

# dropping duplicates
sales_data = sales_data.drop_duplicates()

sales_data
```

```python
# assigning new column names

column_names = ["order_id", "product", "quantity", "price", "date","full_address" ,"address", "city", "state", "zip"]

sales_data.columns = column_names 
```

```python
# changing the datatypes for price and date

sales_data["price"] = sales_data["price"].astype(float)
sales_data["date"] = pd.to_datetime(sales_data["date"])
```

```python
# changing the date to index

# sales_data = sales_data.set_index("date")
```

```python
# What was the best sales month? What was the revenue for that month? Draw a chart showing sales for each month
monthly_sales = sales_data.resample('M', on='date').sum()
```

```python
# Highest Sales happened in the Month of December, and the total sales or revenue is 4594118.95 $. 

# Normal Plot
monthly_sales.plot()
```

```python
# Altair Chart
# resetting index
monthly_sales = monthly_sales.reset_index(inplace= False)

alt.Chart(monthly_sales).mark_line().encode(
    x='date',
    y='price'
)
```

```python
# What US city has the highest number of sales?

# -----
# San Francisco

city_sales = sales_data.groupby("city")["price"].sum()
city_sales = city_sales.reset_index(inplace= False)

#Altair Chart

alt.Chart(city_sales).mark_bar(size=30).encode(
    x='city',
    y='price',
    color = "city"
).properties(width = 500)
```

```python
# we want to buy paid advertisements on the website to promote our products. What is the best time to maximize the 
# likelihood of customerâ€™s responding to the advertisement and purchasing the product?

# -----
# There is growth in sales of the products since October until December, That is the time period which would ideally,
# translate advertisements to purchases
```

```python
# What products are most often bought together by the customer?

# grouping the dataset by date and order_id

# 

items = sales_data.groupby(['date','order_id']).agg({'product': lambda x: x.ravel().tolist()}).reset_index()

items["product_len"] = items['product'].str.len()

# # most brought together Items

items = items.sort_values(by="product_len", ascending = False)

# function to calculate the actual counts of products
def to_1D(series):
 return pd.Series([x for _list in series for x in _list])


# Most purchased products
to_1D(items['product']).value_counts()
```
